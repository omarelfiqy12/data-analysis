```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"fake_customer_data_with_errors (1).xls.csv")
```

```python
df.head()
```

```python
# فصل العمود الوحيد
df = df.iloc[:, 0].str.split(",", expand=True)
print(df.head())
```

```python
# أول صف = أسماء العمدة
df.columns = [
    "CustomerID",
    "Name",
    "Age",
    "Gender",
    "Email",
    "Phone",
    "PurchaseAmount",
    "PurchaseDate",
    "ProductCategory",
    "Rating","unnamed","gander"]

```

```python

print(df.shape)
df.head()
```

```python
df=df.drop(columns=['gander','unnamed','Phone','Email'])
print(df.shape)
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
df['Gender'].value_counts(dropna=False)
```

```python
df["Gender"]=df["Gender"].replace({'M':'Male','male':'Male','F':'Female','female':'Female'})
```

```python
df["Gender"]=df["Gender"].fillna('Male')
```

```python
df["Gender"].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Age'].dtype
```

```python
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
```

```python
df['Age']=df['Age'].apply( lambda x :(df['Age'].mean())if x<0 else x)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df['PurchaseAmount'].dtype
```

```python
df['PurchaseAmount']=pd.to_numeric(df['PurchaseAmount'],errors='coerce')
```

```python
df['PurchaseAmount']=df['PurchaseAmount'].fillna(df['PurchaseAmount'].mean())
```

```python
df['ProductCategory']=df['ProductCategory'].fillna('Unknow')
```

```python
df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
df.isnull().sum()
```

```python
df['Name'].value_counts(dropna=False)
```

```python
df.isnull().sum()
```

```python
x=df.groupby('ProductCategory')['PurchaseAmount'].mean()
x.plot(kind='bar',title='Avg Sales by Product Category')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
x=df.groupby('ProductCategory')['PurchaseAmount'].sum()
x.plot(kind='pie',title='Total Sales by Product Category', autopct='%1.1f%%')
```

```python
x=df.groupby('ProductCategory')['PurchaseAmount'].count()
x.plot(kind='barh',title='Number of Orders by Product Category')
plt.show()
```
