```python
y=pd.merge(df3,df1,on='date',how='left')
```

```python
df=df.drop_duplicates()
```

```python
Total_Revenue_per_Month = df['monthly_fee'].sum()
x=f'${Total_Revenue_per_Month:,.2f}'
print(x)
```

```python
Average_Revenue_per_Month=df['monthly_fee'].mean()
x=f'${Average_Revenue_per_Month:,.2f}'
print(x)
```

```python
Customer_Count_per_Plan=df.groupby('plan_name')['plan_id'].count()
x=Customer_Count_per_Plan.max()
y=Customer_Count_per_Plan.idxmax()
s=f'{y}:{x}'
print(s)
```

```python
total_sms_per_plan = df.groupby('plan_name')['sms'].sum()
print(total_sms_per_plan)
```

```python
arpu_per_plan = df.groupby('plan_name')['monthly_fee'].mean()
print(arpu_per_plan)
```

```python
Customer_Count_per_Plan=df.groupby('plan_name')['plan_id'].count()
Customer_Count_per_Plan.plot(kind='bar',title='Customer Count per Plan')
plt.show()
```

```python
avg_data_per_plan = df.groupby('plan_name')['data_gb'].mean()
avg_data_per_plan.plot(kind='line',title='avg data per plan')
plt.xticks(rotation=90)
plt.show()
```

```python
x=df['plan_name'].value_counts()
x.plot(kind='pie',autopct='%1.1f%%',title='Distribution of Plan Names')
plt.show()
```

```python
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution of Customers')
plt.show()
```

```python
plan_avg = df.groupby('plan_name')[['data_gb','minutes','sms']].mean()
plan_avg.plot(kind='bar', stacked=True)
plt.title('Average Data/Minutes/SMS per Plan')
plt.xticks(rotation=45)
plt.show()
```
